```python
 import pandas as pd
```

```python
df = pd.read_csv("houses_to_rent.csv")
```

```python
df.head()
```

```python
#check nulls
df.info()
```

```python
#removing 'Sem info' and 'Incluso'
df['hoa'] = df['hoa'].replace(['Sem info'], '0')
df['hoa'] = df['hoa'].replace(['Incluso'], '0')

df['rent amount'] = df['rent amount'].replace(['Sem info'], '0')
df['rent amount'] = df['rent amount'].replace(['Incluso'], '0')

df['property tax'] = df['property tax'].replace(['Sem info'], '0')
df['property tax'] = df['property tax'].replace(['Incluso'], '0')
```

```python
#there is not nulls values

# Transform `total` data type from object to integer
df["total"] = (
    df["total"]
    .str.replace("$","",regex=False)
    .str.replace(",","")
    .str.replace("R","")
    .astype(int)
)

# Transform `fire insurance` data type from object to integer
df["fire insurance"] = (
    df["fire insurance"]
    .str.replace("$","",regex=False)
    .str.replace(",","")
    .str.replace("R","")
    .astype(int)
)

# Transform `property tax` data type from object to integer
df["property tax"] = (
    df["property tax"]
    .str.replace("$","",regex=False)
    .str.replace(",","")
    .str.replace("R","")
    .astype(int)
)

# Transform `rent amount` data type from object to integer
df["rent amount"] = (
    df["rent amount"]
    .str.replace("$","",regex=False)
    .str.replace(",","")
    .str.replace("R","")
    .astype(int)
)

# Transform `hoa` data type from object to integer
df["hoa"] = (
    df["hoa"]
    .str.replace("$","",regex=False)
    .str.replace(",","")
    .str.replace("R","")
    .astype(int)
)

df["floor"] = (
    df["floor"]
    .str.replace("-","0")
)

df.head()
```

```python
df.loc[df['furniture'] == 'furnished', 'furniture',] = 1
df.loc[df['furniture'] == 'not furnished', 'furniture',] = 0

df.loc[df['animal'] == 'acept', 'animal',] = 1
df.loc[df['animal'] == 'not acept', 'animal',] = 0
```

```python
df.head()
```

```python
#changing first column name
df.rename(columns={'Unnamed: 0':'house number'},inplace = True)
```

```python
df.head()
```

```python
#droping the duplicates
df.drop_duplicates(inplace = True)
df.shape
#there is no duplicates
```
